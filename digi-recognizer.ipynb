```python
train = pd.read_csv('/kaggle/input/digit-recognizer/train.csv')
train = train.values
train[:5]
```

```python
print(train.shape[0])
print(train.shape[1])
```

```python
X_train = train[:, 1:]
y = train[:, 0]

X_train[:5]
```

```python
X_train = X_train / 255 #scaling
```

```python
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(X_train, y, test_size=0.15, random_state=42)
```

```python
print(X_tr.shape, X_val.shape)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LeakyReLU
from tensorflow.keras.activations import linear, relu, sigmoid, softmax
```

```python
model = Sequential (
[
    Dense(25, activation='relu'),
    Dense(15, activation='relu'),
    Dense(10, activation='softmax')
]

)
```

```python
model.compile(
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    metrics=['accuracy']
)
```

```python
model.fit(X_tr,y_tr, epochs=13, validation_data=(X_val,y_val))
```

```python
val_loss, val_acc = model.evaluate(X_val, y_val, verbose=0)
print(val_acc)
```

```python
test = pd.read_csv('/kaggle/input/digit-recognizer/test.csv')
test = test /255

pred = model.predict(test, verbose=1)


y_pred = pred.argmax(axis=1)
```

```python
submission = pd.DataFrame({
    "ImageId": np.arange(1, len(y_pred)+1),
    "Label": y_pred
})

submission.to_csv("submission.csv", index=False)
```
